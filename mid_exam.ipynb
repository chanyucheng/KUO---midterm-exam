```python
# 筆記 判斷符號的方法
def printOneRow(m, n, s, mark):
    for y in range(m, n, s): 
        if mark.isdigit():
            print(y, end='') 
        else:
            print(mark, end='') 
def myPrint1(n):
    for x in range(0, n):
        printOneRow(0, x, 1, '#')
        printOneRow(2*(n-x-1), -1, -2, '0')
        printOneRow(2, 2*(n-x), 2, '0')        
        print()
myPrint1(4)
```

```python
# 第一題 : 圖形
# 心得 : 如果單行沒有迭代那一層迴圈即可

# 關鍵函式 p1r()
def p1r(m,n,s,mark):
    for y in range(m,n,s):
        if mark.isdigit():
            print(y,end="")
        else:
            print(mark,end="")


n = int(input())
l = int(input())

def func_1(l):
    for i in range(1,l+1):
        print(f"{i}"*i,end="")
        print("#"*(l-i),end="")
        print()
# func_1(l)

def func_2(l):
    for i in range(l):
        p1r(2*(l-i-1),0,-1,'#')
        p1r(1,i+2,1,'0')
        p1r(i,0,-1,'0')
        print()
# func_2(l)

def func_3(l):
    for i in range(l):
        p1r(1,i+2,1,'0')
        p1r((l-1-i),0,-1,'^')
        print()
    for j in range(l):
        p1r((l-j),0,-1,'0')
        p1r(1,j+1,1,'^')
        print()
# func_3(l)

def func_4(l):
    for i in range(l):
        p1r((l-1-i),0,-1,'^')
        p1r(i+1,0,-1,'0')
        p1r(2,i+2,1,'0')
        p1r((l-1-i),0,-1,'^')
        print()
    for j in range(l-1):
        p1r(0,j+1,1,'^')
        p1r((l-1-j),0,-1,'0')
        p1r(2,(l-j),1,'0')
        p1r(0,j+1,1,'^')
        print()
# func_4(l)

if n==1:
    func_1(l)
elif (n==2):
    func_2(l)
elif (n==3):
    func_3(l)
elif(n==4):
    func_4(l)
```

```python
# 課程衝堂 
def check_conflicts(course_total, course_data):
    # 檢查課程總數是否在範圍內
    if not (2 <= course_total <= 10):
        return -1

    # 儲存課程資料
    a = course_data

    # 檢查每個課程的時間是否符合格式規範
    for u in range(len(a)):
        if not (1 <= len(a[u][1]) <= 3):
            return -1
        for k in range(len(a[u][1])):
            # 將時間轉換為字串
            time_str = str(a[u][1][k])
            
            # 檢查長度是否為 2
            if len(time_str) != 2:
                return -1
            
            # 檢查第一位是否在 1-5 之間
            if not ('1' <= time_str[0] <= '5'):
                return -1
            
            # 檢查第二位是否是 1-9 或 a, b, c
            if not (time_str[1].isdigit() and '1' <= time_str[1] <= '9') and time_str[1] not in ['a', 'b', 'c']:
                return -1

    # 儲存每個課程與其時間的對應
    courses = []
    for i in range(len(a)):
        for j in range(len(a[i][1])):
            courses.append((a[i][0], a[i][1][j]))

    # 儲存衝堂的結果
    conflicts = []

    # 檢查課程是否衝堂
    for h in range(len(courses)):
        for k in range(h + 1, len(courses)):
            if courses[h][1] == courses[k][1]:
                conflicts.append((courses[h][0], courses[k][0], courses[h][1]))

    # 如果沒有衝堂，返回 "correct"
    if not conflicts:
        return "correct"
    else:
        # 返回衝堂資訊
        return [f"{conflict[0]} and {conflict[1]} conflict on {conflict[2]}" for conflict in conflicts]


# 測試函式
course_total = int(input())  # 選出幾門課
course_data = [[input(), [input() for i in range(int(input()))]] for z in range(course_total)]

# 呼叫函式並取得結果
result = check_conflicts(course_total, course_data)

# 輸出結果
if result == -1:
    print(-1)
elif result == "correct":
    print("correct")
else:
    for conflict in result:
        print(conflict)
```
